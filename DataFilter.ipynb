```python
import pandas as pd
import os
import glob

```

```python
# Define the input and output directories
input_directory = r'C:\Users\Milan\OneDrive\Desktop\New folder\gdelt_data\extracted'
output_directory = r'C:\Users\Milan\OneDrive\Desktop\New folder\new'

```

```python
# Ensure the output directory exists
os.makedirs(output_directory, exist_ok=True)

```

```python
# Get a list of all CSV files in the input directory
csv_files = glob.glob(os.path.join(input_directory, '*.CSV'))

```

```python
# Process each CSV file
for file_path in csv_files:
    # Load the data
    gdelt_data = pd.read_csv(file_path, delimiter='\t', header=None)
    
    # Filter the data for Nepal
    nepal_data = gdelt_data[gdelt_data[51] == 'NP']
    
    # Construct the output file path
    file_name = os.path.basename(file_path)
    output_file_path = os.path.join(output_directory, file_name)
    
    # Save the Nepal data to a new CSV file
    nepal_data.to_csv(output_file_path, index=False, header=False)

print("Nepal-related data extraction completed for all files.")
```
